```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
mobilenet_freeze_cambodge = pd.read_csv("mobilenet_freeze_cambodge_results.csv")
mobilenet_freeze_guinee = pd.read_csv("mobilenet_freeze_guinee_results.csv")
```

#### Results of mobilenet_v2 on cambodge

```python
print(f'Mean height relative error = {mobilenet_freeze_cambodge.height_relative_error.mean()}')
print(f'Mean weight relative error = {mobilenet_freeze_cambodge.weight_relative_error.mean()}')
mobilenet_freeze_cambodge.head()
```

```python

## show height error
plt.figure(figsize=(4, 3))
plt.scatter(mobilenet_freeze_cambodge.true_height,mobilenet_freeze_cambodge.height_relative_error)
plt.xlabel("height")
plt.ylabel('relative height error')
plt.hlines(y=0.05,xmin=45,xmax=110,colors='r',label='5% error')

## show weight error
plt.figure(figsize=(4, 3))
plt.scatter(mobilenet_freeze_cambodge.true_weight,mobilenet_freeze_cambodge.weight_relative_error)
plt.xlabel("weight")
plt.ylabel('relative weight error')
plt.hlines(y=0.05,xmin=2,xmax=25,colors='r',label='5% error')
```

#### Results of mobilenet_v2 on Guinee

```python
print(f'Mean height relative error = {mobilenet_freeze_guinee.height_relative_error.mean()}')
print(f'Mean weight relative error = {mobilenet_freeze_guinee.weight_relative_error.mean()}')
mobilenet_freeze_guinee.head()
```

```python

## show height error
plt.figure(figsize=(4, 3))
plt.scatter(mobilenet_freeze_guinee.true_height,mobilenet_freeze_guinee.height_relative_error)
plt.xlabel("height")
plt.ylabel('relative height error')
plt.hlines(y=0.05,xmin=40,xmax=60,colors='r',label='5% error')

## show weight error
plt.figure(figsize=(4, 3))
plt.scatter(mobilenet_freeze_guinee.true_weight,mobilenet_freeze_guinee.weight_relative_error)
plt.xlabel("weight")
plt.ylabel('relative weight error')
plt.hlines(y=0.05,xmin=0,xmax=6,colors='r',label='5% error')
```
