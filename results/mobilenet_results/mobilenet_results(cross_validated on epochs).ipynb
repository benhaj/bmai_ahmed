```python
import glob
listed = glob.glob("*.csv")
dfs = []
for i in range(len(listed)):
    print(30*'=')
    elem = listed[i]
    df = pd.read_csv(elem)
    dfs.append(df)
    mean_h = df.height_relative_error.mean()
    mean_w = df.weight_relative_error.mean()
    print(f'{elem} have results : ')
    print(20*'-')
    print('mean relative errors:')
    print(f'        height: {mean_h}')
    print(f'        weight: {mean_w}')    
```
